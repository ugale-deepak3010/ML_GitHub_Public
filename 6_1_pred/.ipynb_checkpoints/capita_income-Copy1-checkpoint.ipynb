```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data_frame__df=pd.read_csv("canada_per_capita_income.csv")
data_frame__df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(data_frame__df.year,data_frame__df.per_capita_income,color='red',marker='+')
```

```python
new_df = data_frame__df.drop('per_capita_income',axis='columns')
new_df
```

```python
new_per_capita_income_df = data_frame__df.drop('year',axis='columns')
new_per_capita_income_df
```

```python
deep_learning_model_regression = linear_model.LinearRegression()
deep_learning_model_regression.fit(new_df,new_per_capita_income_df)
```

```python
predict=deep_learning_model_regression.predict([[2020]])
predict
```
